```python

def calculate_metrics(df):

    df['SalesRevenue'] = df['Quantity'] * df['Price']

    # Total Sales Revenue
    total_sales_revenue = df['SalesRevenue'].sum()

    # Average Order Value (AOV)
    total_orders = len(df)
    aov = total_sales_revenue / total_orders

    # Total Quantity Sold
    total_quantity_sold = df['Quantity'].sum()

    # Sales by Product
    sales_by_product = df.groupby('Itemname').agg({
        'SalesRevenue': 'sum',
        'Quantity': 'sum'
    }).reset_index()

    # Sales by Country
    sales_by_country = df.groupby('Country').agg({
        'SalesRevenue': 'sum',
        'Quantity': 'sum'
    }).reset_index()

    # Customer Purchase Frequency
    purchase_frequency = df.groupby('CustomerID').agg({
        'Date': 'count'
    }).reset_index()
    purchase_frequency.rename(columns={'Date': 'PurchaseCount'}, inplace=True)

    # Customer Total Spending
    customer_spending = df.groupby('CustomerID').agg({
        'SalesRevenue': 'sum'
    }).reset_index()


    metrics = {
        'Total Sales Revenue': total_sales_revenue,
        'Average Order Value (AOV)': aov,
        'Total Quantity Sold': total_quantity_sold,
        'Sales by Product': sales_by_product,
        'Sales by Country': sales_by_country,
        'Customer Purchase Frequency': purchase_frequency,
        'Customer Total Spending': customer_spending
    }

    return metrics
```

```python

usa_metrics = calculate_metrics(df_usa_clean)
others_metrics = calculate_metrics(df_others_clean)
df_overall_clean = pd.concat([df_usa_clean, df_others_clean], ignore_index=True)
overall_metrics = calculate_metrics(df_overall_clean)
print("USA Metrics:")
print(f"Total Sales Revenue : {usa_metrics['Total Sales Revenue']}")
print(f"Average Order Value (AOV) : {usa_metrics['Average Order Value (AOV)']}")
print(f"Total Quantity Sold : {usa_metrics['Total Quantity Sold']}")
print(f"Sales By Product : {usa_metrics['Sales by Product']}")
print(f"Customer Purchase Frequency : {usa_metrics['Customer Purchase Frequency']}")
print(f"Customer Total Spending : {usa_metrics['Customer Total Spending']}")


print("\nOther Countries Metrics:")
print(f"Total Sales Revenue : {others_metrics['Total Sales Revenue']}")
print(f"Total Quantity Sold : {others_metrics['Total Quantity Sold']}")
print(f"Sales By Product : {others_metrics['Sales by Product']}")
print(f"Customer Purchase Frequency : {others_metrics['Customer Purchase Frequency']}")
print(f"Customer Total Spending : {others_metrics['Customer Total Spending']}")

print("\nOverall Metrics:")
print(f"Total Sales Revenue : {overall_metrics['Total Sales Revenue']}")
print(f"Average Order Value (AOV) : {overall_metrics['Average Order Value (AOV)']}")
print(f"Total Quantity Sold : {overall_metrics['Total Quantity Sold']}")
print(f"Sales By Country : {overall_metrics['Sales by Country']}")
print(f"Sales By Product : {overall_metrics['Sales by Product']}")
print(f"Customer Purchase Frequency : {overall_metrics['Customer Purchase Frequency']}")
print(f"Customer Total Spending : {overall_metrics['Customer Total Spending']}")

```

Sales Trend Analysis

```python
df_usa_clean['SalesRevenue'] = df_usa_clean['Quantity'] * df_usa_clean['Price']
df_others_clean['SalesRevenue'] = df_others_clean['Quantity'] * df_others_clean['Price']
df_overall_clean['SalesRevenue'] = df_overall_clean['Quantity'] * df_overall_clean['Price']
df_usa_sales = df_usa_clean[['Date', 'SalesRevenue']]
df_others_sales = df_others_clean[['Date', 'SalesRevenue']]
df_overall_sales = df_overall_clean[['Date', 'SalesRevenue']]
df_usa_sales['Date'] = pd.to_datetime(df_usa_sales['Date'], errors='coerce')
df_others_sales['Date'] = pd.to_datetime(df_others_sales['Date'], errors='coerce')
df_overall_sales['Date'] = pd.to_datetime(df_overall_sales['Date'], errors='coerce')

df_usa_sales.set_index('Date', inplace=True)
df_others_sales.set_index('Date', inplace=True)
df_overall_sales.set_index('Date', inplace=True)

usa_sales_trend = df_usa_sales.resample('M').sum()
others_sales_trend = df_others_sales.resample('M').sum()
overall_sales_trend = df_overall_sales.resample('M').sum()
#for plots
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(usa_sales_trend.index, usa_sales_trend['SalesRevenue'], marker='o', color='blue', label='USA Sales Revenue')
plt.title('USA Sales Trend (Monthly)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.subplot(3, 1, 2)
plt.plot(others_sales_trend.index, others_sales_trend['SalesRevenue'], marker='o', color='green', label='Others Sales Revenue')
plt.title('Others Sales Trend (Monthly)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.subplot(3, 1, 3)
plt.plot(overall_sales_trend.index, overall_sales_trend['SalesRevenue'], marker='o', color='red', label='Overall Sales Revenue')
plt.title('Overall Sales Trend (Monthly)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
